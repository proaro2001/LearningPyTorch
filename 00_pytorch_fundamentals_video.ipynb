####***Two Main rules to satisified to perform matrix multiplication***
1. The **inner dimensions** must match:
* `@` here is just `matmul`
* `(3,2) @ (3, 2)` won't work
* `(2,3) @ (3, 2)` will work
* `(3,2) @ (2, 3)` will work
2. The resulting matrix has the shape of the **outer dimensions**
* `(2,3) @ (3,2)` -> (result) `(2,2)`
* `(3,2) @ (2,3)` -> (result) `(3,3)`

### One of the most common errors in deep learning: shape errors

```python
import torch
# Shapes for matrix multiplication
tensor_A = torch.tensor([[1,2],
                         [3,4],
                         [5,6]])

tensor_B = torch.tensor([[7,10],
                         [8,11],
                         [9,12]])
try:
  # torch.mm( tensor_A, tensor_B ) # torch.mm is the same as torch.matmul ( it's an alias for writing matmul)
  torch.matmul( tensor_A, tensor_B )
except Exception as e:
  print (e)
```

```python
tensor_A.shape, tensor_B.shape
```

### transpose

To fix our tensor shape issues, we can manipulate the shape
of one of our tensors using a **transpose**.

A **transpose** switches the axes or dimensions of a given tensor


```python
tensor_B, tensor_B.shape
```

```python
# T stands for tenspose
tensor_B.T, tensor_B.T.shape
```

```python
torch.matmul(tensor_A,tensor_B.T)
```

```python
torch.matmul(tensor_A,tensor_B.T).shape
```

```python
# The matrix multiplication operation works when tensor_B is transposed
print( f"Origional shapes: tensor_A = { tensor_A.shape}, tensor_B = {tensor_B.shape}")
print( f"New shapes: tesnor_A = {tensor_A.shape} (same shape as above), tensor_B.T = {tensor_B.T.shape}")
print( f"Multiplying: {tensor_A.shape} @ {tensor_B.T.shape } <- inner dimensions must match")
print( "Output:\n")
output = torch.matmul ( tensor_A, tensor_B.T)
print(output)
print( f"n Output shape: {output.shape}")

```

### Reshaping, stacking, squeezing and unsqueezing tensors


---

* Reshaping - reshapes an input tensor to a defined shape
* View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking - combine multiple tensors on top of each other (vstack , vertical stack) or side by side (hstack, horizontal stack)
* Squeeze - remove all `1` dimensions from a tensor
* Unsqueeze - add a `1` dimension to a target tensor
* Permute - Return a view of the input with dimensions permuted (swapped) in a certain way


```python
# Let's create a tensor
x = torch.arange(1.,10.)
x, x.shape
```

```python
# Add an extra dimension
# Note enough space for 9 elements
try:
  x_reshaped = x.reshape(1, 7)
  x_reshaped, x_reshaped.shape
except Exception as e:
  print(e)
```

```python
# Add an extra dimension
# valid space
try:
  x_reshaped = x.reshape(1, 9)
except Exception as e:
  print(e)

x_reshaped, x_reshaped.shape
```

```python
# Add an extra dimension
# leaving too much space for the elements
try:
  x_reshaped = x.reshape(2, 9)
  x_reshaped, x_reshaped.shape
except Exception as e:
  print(e)
```

```python
# Add an extra dimension
# leaving too much space for the elements
try:
  x_reshaped = x.reshape(9, 1)
except Exception as e:
  print(e)


x_reshaped, x_reshaped.shape
```

```python
# Let's create a tensor
x = torch.arange(1.,11.)
x, x.shape
```

```python
# Add an extra dimension
# leaving too much space for the elements
try:
  x_reshaped = x.reshape(5,2)
except Exception as e:
  print(e)


x_reshaped, x_reshaped.shape
```

```python
# torch.permute - rearranges the dimensions of a target tensor in a specific order
x_origional = torch.rand( size=(224, 224, 3) ) # [height, width, coloer_channels]

# Permute the origional tensor to rearrange the axis ( or dim ) order
x_permuted = x_origional.permute(2, 0, 1) # shifts axis 0->1, 1->2, 2->0

print( f"Previous shape: {x_origional.shape}")
print( f"New shape: {x_permuted.shape}") # [color_cahnnels, height, width]
```

### Reproducbility ( trying to take random out of random )

### To reduce the randomness in neural networks and PyTorch comes the concept of a **random seed**

---



```python
import torch

# Create two random tensors
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
# Let's make some andom but reproducible tensors
import torch

# Set the random seed
RANDOM_SEED = 42
torch.manual_seed( RANDOM_SEED )
random_tensor_C = torch.rand(3,4)

torch.manual_seed( RANDOM_SEED )
random_tensor_D = torch.rand(3,4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
